### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
data = requests.get("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
dataCont = data.content
dataContAcc = BeautifulSoup(dataCont, "html5lib")
dataContAcc
# soup = BeautifulSoup(dataCont, 'lxml')
# print(soup.prettify())

# url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
# chipo = pd.read_csv(url, sep = '\t')
```

```python
body = dataContAcc.select("body")
```

```python
bodyList = str(body[0]).split("\n")
```

```python
header = bodyList[0].split("\t")
header[0] = 'order_id'
header
```

```python
dataList = []
for i in bodyList[1:]:
  dataList.append(i.split("\t"))
del dataList[-1]
dataList[:5]
```

```python
df = pd.DataFrame(dataList, columns = header)
df
```
