### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep = "\t")
print(chipo.shape[0])
chipo.head(2)
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].replace('$', '')
chipo['item_price'] = chipo['item_price'].astype(float)

chipo[chipo['item_price'] > 10.00].count()[0]
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
l = []
lp = []
for i in range(len(chipo)):
  if chipo['item_name'][i] not in l:
    if chipo['quantity'][i] == 1:
      l.append(chipo['item_name'][i])
      lp.append(chipo['item_price'][i])

l = np.array(l)
lp = np.array(lp, dtype=float)
df = pd.DataFrame({"item_name": l, "item_price": lp}, columns = ["item_name","item_price"])
df
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by=['item_name']) # (.reset_index())
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# xx = np.where(chipo.item_price == chipo.item_price.max(),chipo.quantity,0)
# for i in xx:
#     if i!=0:
#         print("Quantity of the most expensive item ordered is : ",i)

chipo['quantity'][chipo['item_price'] == chipo['item_price'].max()].values[0]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
c = 0
cq = 0
for i in range(len(chipo)):
  if chipo['item_name'][i] == "Veggie Salad Bowl":
    c += 1
    cq += 1 * chipo['quantity'][i]
print(c, cq)

chipo[chipo['item_name'] == "Veggie Salad Bowl"].count()[0]
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
cs = 0
for i in range(len(chipo)):
  if chipo['item_name'][i] == "Canned Soda":
    if chipo['quantity'][i] > 1:
      cs += 1
print(cs)

chipo[(chipo['item_name'] == "Canned Soda") & (chipo['quantity'] > 1)].count()[0]
```
